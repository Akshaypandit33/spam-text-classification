```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv", encoding='latin-1')
data.head()
```

```python
data=data.iloc[:,:2]
# to change the column name
data.columns=['label', 'text']
data.head()
```

```python
# to check missing value

data.isna().sum()
```

```python
# text preprocessing

#downld nltk (natural language tool kit)
import nltk
nltk.download('all')

```

```python
#create a list text
text= list(data['text'])

#preprocessing loop
import re

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer= WordNetLemmatizer()


corpus=[]

for i in range(len(text)):
  r= re.sub('[^a-zA-Z]', ' ', text[i])
  r= r.lower()
  r=r.split()

  r=[word for word in r if word not in stopwords.words('english')]
  r=[lemmatizer.lemmatize(word) for word in r]

  r = ' '.join(r)

  corpus.append(r)


data['text'] = corpus

data.head()
```

```python
#create feature and label sets
x=data['text']
# target variable
y=data['label']


#traoin test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.33, random_state=123)


print('Training Data :', x_train.shape)
print('Testing Data :', x_test.shape)
```

```python

#Tarin Bag of Words model feature extraction

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()

x_train_cv=cv.fit_transform(x_train)
x_train_cv.shape
```

```python
# Training Logistic Regression model
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train_cv,y_train)


# transform x_test using cv
x_test_cv= cv.transform(x_test)
```

```python
#generate predictions

predictions = lr.predict(x_test_cv)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test,predictions)
```
